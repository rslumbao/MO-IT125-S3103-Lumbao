```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('C:/Users/rlumbao/Downloads/customerdemo.csv')
ax = sns.histplot(df['Age'], kde=True, bins=10, color='skyblue')
for bar in ax.patches:
    bar_height = bar.get_height()  # Get the height of the bar (frequency)
    ax.text(
        bar.get_x() + bar.get_width() / 2,  # X-coordinate: center of the bar
        bar_height + 0.5,  # Y-coordinate: slightly above the bar
        f'{int(bar_height)}',  # Display the height as an integer
        ha='center', va='bottom', fontsize=10, color='black'
    )
plt.title('Age Distribution', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('C:/Users/rlumbao/Downloads/customerdemo.csv')
ax = sns.countplot(data=df, x='Gender', hue='Gender', dodge=False, palette='pastel', legend=False)
for bar in ax.patches:
    bar_height = bar.get_height()  # Get the height of the bar
    ax.text(
        bar.get_x() + bar.get_width() / 2,  # X-coordinate of the label
        bar_height + 0.5,  # Y-coordinate slightly above the bar
        f'{int(bar_height)}',  # Label text (convert to int)
        ha='center', va='bottom', fontsize=10, color='black'
    )
plt.title('Gender Distribution', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('C:/Users/rlumbao/Downloads/customerdemo.csv')
ax = sns.boxplot(x='IncomeLevel', y='Age', hue='Gender', data=df, palette='pastel')
for i, box in enumerate(ax.artists):
    median = df[(df['IncomeLevel'] == ax.get_xticklabels()[i//2].get_text()) & (df['Gender'] == ax.get_legend_handles_labels()[0][i % 2])]['Age'].median()
    ax.text(
        box.get_x() + box.get_width() / 2,  # X coordinate (center of the box)
        median,  # Y coordinate (median value)
        f'{median:.2f}',  # Display median to 2 decimal places
        ha='center', va='center', fontsize=10, color='black'
    )
counts = df.groupby(['IncomeLevel', 'Gender'])['Age'].count().reset_index(name='Count')
for i, (income, gender) in enumerate(counts[['IncomeLevel', 'Gender']].values):
    count = counts.loc[(counts['IncomeLevel'] == income) & (counts['Gender'] == gender), 'Count'].values[0]
    ax.text(
        i % len(ax.get_xticklabels()),  # X position (adjust to match positions of income level)
        0.95,  # Y position, slightly above the box
        f'Count: {count}',  # Display count
        ha='center', va='bottom', fontsize=10, color='black'
    )
plt.title('Income Level by Gender and Age', fontsize=16)
plt.xlabel('Income Level', fontsize=12)
plt.ylabel('Age', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
